### 3.
Дополните код Монте-Карло последовательности независимых испытаний расчетом соответствующих вероятностей (через биномиальное распределение) 
и сравните результаты.


```python
import math
k, n = 0, 10000
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)
x = a + b + c + d
K = 2
N = 4
for i in range(0, n):
    if x[i] == K:
        k = k + 1
print (k, n, k/n)
Cnk = math.factorial(N) / (math.factorial(K) * (math.factorial(N - K)))
Pnk = Cnk/2**N
print (Pnk)
```

0.3702 и 0.375 очень близки

Повторите расчеты биномиальных коэффициентов и вероятностей k успехов в последовательности из n независимых испытаний, взяв другие значения n и k.

```python
k, n = 0, 10000
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)
e = np.random.randint(0, 2, n)
f = np.random.randint(0, 2, n)
x = a + b + c + d + e + f
K = 4
N = 6
for i in range(0, n):
    if x[i] == K:
        k = k + 1
print (k, n, k/n)
Cnk = math.factorial(N) / (math.factorial(K) * (math.factorial(N - K)))
Pnk = Cnk/2**N
print (Pnk)
```

### 4. Из урока по комбинаторике повторите расчеты, сгенерировав возможные варианты перестановок для других значений n и k


```python
import itertools
for p in itertools.product('0123456', repeat=4):
    print(''.join(p))
```

```python
for p in itertools.permutations('012', 3):
    print(''.join(p))
```

```python
for p in itertools.combinations('0123456', 5):
    print(''.join(p))
```
