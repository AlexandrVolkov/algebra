### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,



```python
x = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(x), np.sin(x))
plt.gca().set_aspect(1)
```

эллипса,


```python
x = np.linspace(0, 2*np.pi, 100)
plt.plot(2 * np.cos(x), 1.5 * np.sin(x))
plt.gca().set_aspect(1)
```

гиперболы.

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, 1/x)
plt.gca().set_aspect(1)
```

### 5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.


```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X,Y = np.meshgrid(X, Y)
Z = 2*X +5*Y + 15
ax.plot_wireframe(X, Y, Z)
Z = 2*X +5*Y - 15
ax.plot_wireframe(X, Y, Z)
show()
```

Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X,Y = np.meshgrid(X, Y)
Z = X**2 + Y**2
ax.plot_wireframe(X, Y, Z)
fig2 = figure()
ax2 = Axes3D(fig2)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X,Y = np.meshgrid(X, Y)
Z = np.sqrt(1 * X**2 + 1 * Y**2)
ax2.plot_wireframe(X, Y, Z)
show()
```
